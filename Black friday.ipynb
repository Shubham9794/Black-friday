```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# import train data
df_train=pd.read_csv(r"C:\Users\shubh\Downloads\Telegram Desktop\train.csv")
df_train.head()
```

```python
# import test data
df_test=pd.read_csv(r"C:\Users\shubh\Downloads\Telegram Desktop\test.csv")
df_test.head()
```

```python
# merge both data 
df=pd.concat([df_train,df_test],axis=0)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.replace(to_replace=['F','M'],value=[0,1],inplace=True)
df.head()
```

```python
# handling categorical features age 
df['Age'].unique()
```

```python
# label encoding for Age 
df['Age']=df['Age'].map({"0-17":1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df.head()
```

```python
# label encoding for city 
df_city=pd.get_dummies(df['City_Category'],drop_first=True, dtype=int)
df_city
```

```python
df=pd.concat([df,df_city],axis=1)
df.head()
```

```python
# drop city category
df.drop('City_Category',axis=1,inplace=True)
df.head()
```
